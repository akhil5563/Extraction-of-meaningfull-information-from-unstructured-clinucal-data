```python
import spacy
import nltk
import os
import re
import string
import random

nlp = spacy.load('en_core_web_sm')

dirt=[]
path=r"C:\Users\91955\Desktop\Data"
os.chdir(path)       
for file in os.listdir():
    
    if file.endswith(".txt"):
        
        p=f"{path}\{file}"
        dirt.append(p)
        

m = random.randint(1,len(dirt))
n=dirt[m]
with open (n,'r') as f:
            text1=f.read()

    
text=text1




```

```python
text=re.sub('<.*>','',text)
text=re.sub('ENDOFARTICLE.','',text)
punctuation="["+re.sub("\.","",string.punctuation)+"]"
text = re.sub('\n',' ',str(text))
text=re.sub(' +',' ', text)
text=re.sub(punctuation,"",text)
text=re.sub("\d+", "", text)
    
```

```python
doc = nlp(text)
l=[]

for token in doc:
    l.append(token.text)
l=[i for i in l if i.strip()]
print(l)
```

```python
from nltk.corpus import stopwords 

stop_words=[]
stop = set(stopwords.words('english'))
for i in stop:
    k=i.lower()
    stop_words.append(k)
print(stop_words)
```

```python
filtered_words = [] 
for w in l: 
    if w.lower() not in stop_words: 
        filtered_words.append(w)
for i in filtered_words:
    if len(i)<=1:
        filtered_words.remove(i)
        



print(filtered_words)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.util import ngrams
from collections import Counter

exact1 = WordNetLemmatizer()
lematized_words= [exact1.lemmatize(word) for word in filtered_words]

for i in lematized_words:
    if " " in i:
        lematized_words.remove(i)
 
print(lematized_words)
```

```python
bigrams=ngrams(lematized_words,2)
freqofbigrams=Counter(bigrams)
s=freqofbigrams.most_common(10)
print(s)
```

```python

main_words=list(set(lematized_words))#duplicate words
for i in s:
    m=i[0]
    g=m[0]+' '+m[1]
    main_words.append(g)
print(main_words)



```

```python
#Filtering pos tags
pos_t=nltk.pos_tag(main_words)

print(pos_t)
```

```python
def mapper_f(dict_obj):
    new_dict={}
    for key in dict_obj.keys():
        if dict_obj[key] in new_dict.keys():
            new_dict[dict_obj[key]]=str(new_dict[dict_obj[key]])+','+str(key)
        else:
            new_dict[dict_obj[key]]=key
    return new_dict
dict_obj=dict(pos_t)
p=mapper_f(dict_obj)

print(p)
```

```python
import json

import requests
from lxml.html import fromstring
f=final1

uri="https://utslogin.nlm.nih.gov"
auth_endpoint = "/cas/v1/api-key"

class Authentication:

   def __init__(self, apikey):
    self.apikey=apikey
    self.service="http://umlsks.nlm.nih.gov"

   def gettgt(self):
     params = {'apikey': self.apikey}
     h = {"Content-type": "application/x-www-form-urlencoded", "Accept": "text/plain", "User-Agent":"python" }
     r = requests.post(uri+auth_endpoint,data=params,headers=h)
     response = fromstring(r.text)
     tgt = response.xpath('//form/@action')[0]
     return tgt

   def getst(self,tgt):

     params = {'service': self.service}
     h = {"Content-type": "application/x-www-form-urlencoded", "Accept": "text/plain", "User-Agent":"python" }
     r = requests.post(tgt,data=params,headers=h)
     st = r.text
     return st

p=Authentication("ae301551-c698-4e3b-99f7-cf0ac3c11310")
print("success")
```
